```python
import pandas as pd
data = pd.read_csv('/content/trans.csv',index_col=False) 
# type(data)
data
```

```python
from googletrans import Translator
translator = Translator()
```

```python
data['marathi'] = data['english'].apply(translator.translate, src='en', dest='mr').apply(getattr, args=('text',))
```

```python
data.to_csv('pcl_marathi.csv',encoding = "utf-8-sig")
```
